# Importando as ***Bibliotecas***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('_mpl-gallery')
```

#*Extraindo* os **dados**

```python
df=pd.read_csv('Vendas_Veiculos.csv')
display(df)
```

#Convertendo colunas Objects para ***numeric***

```python
df['Unnamed: 2'] = pd.to_numeric(df['Unnamed: 2'], errors='coerce')
df['Unnamed: 5'] = pd.to_numeric(df['Unnamed: 5'], errors='coerce')
df['Unnamed: 6'] = pd.to_numeric(df['Unnamed: 6'], errors='coerce')
```

```python
df.info()
```

# Soma dos campos ***vazios***

```python
print(df.isna().sum())
```

#Removendo valores ***vazios***



```python
df = df.dropna(how="all", axis=1)
df = df.dropna(how="any", axis=0)
display(df)
```

#Classificando a tabela, pelo o valor do ***carro*** em ordem crescente

```python
df.sort_values(by='Unnamed: 5')
```
